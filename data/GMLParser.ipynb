```python
from collections import defaultdict

import networkx as nx
```

```python
def test(data):
    try:
        return int(data)
    except:
        pass
    try:
        return float(data)
    except:
        pass
    return data

H = nx.read_gml('results/Words_25_topics_10.gml', destringizer=test)
```

```python
dic_modul = nx.get_node_attributes(H, 'ModularityClass')
```

```python
for k, v in dic_modul.items():
    print(k.replace('Posner_', 'Posner topic,').replace('Becker_', 'Becker topic,').replace('Word_', 'Word,') + ',' + str(v))
```

```python
dic_value = nx.get_edge_attributes(H, 'value')
```

```python
strengths = defaultdict(list)
weights = defaultdict(list)
for (source, target), v in dic_value.items():
    m_s = dic_modul[source]
    m_t = dic_modul[target]
    if dic_modul[source] != dic_modul[target]:
        strengths[(min(m_s, m_t), max(m_s, m_t))].append(v)
    else:
        weights[m_s].append(v)
```

```python
for k, v in strengths.items():
    print(k, sum(v), sum(v)/len(v))
```

```python
for k, v in weights.items():
    print(k, sum(v), sum(v)/len(v))
```

```python
import networkx as nx

G = nx.Graph()

for k, v in weights.items():
    G.add_node(f'Class_{k}', weight=sum(v))

for (a, b), v in strengths.items():
    G.add_edge(f'Class_{a}', f'Class_{b}', weight=sum(v))
    
nx.write_gml(G, 'graph.gml')
```

```python
import networkx as nx

G = nx.Graph()

for k, v in weights.items():
    G.add_node(f'Class_{k}', weight=sum(v)/len(v))

for (a, b), v in strengths.items():
    G.add_edge(f'Class_{a}', f'Class_{b}', weight=sum(v)/len(v))
    
nx.write_gml(G, 'graph2.gml')
```
